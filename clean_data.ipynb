```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\satya\OneDrive\Desktop\Take_Home_Assignment_App_dev\export_2019.csv.')
```

```python
df
```

```python
columns_to_clean = [
    'Bottle Volume (ml)',
    'State Bottle Cost',
    'State Bottle Retail',
    'Bottles Sold',
    'Sale (Dollars)',
    'Volume Sold (Liters)',
    'Volume Sold (Gallons)',
]
```

```python
for column in columns_to_clean:
    # Remove commas and convert to numeric
    df[column] = pd.to_numeric(df[column].str.replace(',', '').fillna(0), errors='coerce')
```

```python
df['County Number'] = df['County Number'].apply(lambda x: int(float(x)) if pd.notnull(x) else 0)
```

```python
df.fillna(0, inplace=True)
```

```python
int_columns = ['Category', 'Vendor Number', 'Item Number', 'Pack', 'Store Number']

for column in int_columns:
    df[column] = df[column].apply(lambda x: int(float(x)) if pd.notnull(x) else 0)
```

```python
df.to_csv('cleaned_invoices.csv', index=False)
```
